```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
```

```python
df3=df_origin.copy()
col_X=['SeniorCitizen',
 'Partner',
 'Dependents',
 'tenure',
 'MonthlyCharges',
 'TotalCharges',
 'OnlineSecurity',
 'OnlineBackup',
 'DeviceProtection',
 'TechSupport',
 'StreamingTV',
 'PaperlessBilling']
```

```python
X=df3[col_X].copy()
y=df3['Churn'].copy()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=123)
```

```python
X_train
```

```python
X_test.info()
```

```python
d1=X_train[['Partner','Dependents','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','PaperlessBilling']].applymap(lambda x: 1 if x=='Yes' else (0 if x=='No' else -1))
d2=X_train.select_dtypes(include='number').copy()
X_train=pd.concat([d1,d2],axis=1)
X_train
```

```python
d1=X_test[['Partner','Dependents','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','PaperlessBilling']].applymap(lambda x: 1 if x=='Yes' else (0 if x=='No' else -1))
d2=X_test.select_dtypes(include='number').copy()
X_test=pd.concat([d1,d2],axis=1)
X_test
```

```python
scaler_mm = MinMaxScaler()
```

```python
X_train=scaler_mm.fit_transform(X_train)
```

```python
X_train.shape
```

```python
X_test=scaler_mm.transform(X_test)
```

```python
model_lr = LogisticRegression(random_state = 123)
model_lr.fit(X = X_train,
             y = y_train)
```

```python
y_pred=model_lr.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```
